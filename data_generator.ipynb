```python
from sklearn import datasets

# Cargar el dataset de dígitos
digits = datasets.load_digits()

# Acceder a los datos y etiquetas
X = digits.data  # Las características (imágenes aplanadas)
y = digits.target  # Las etiquetas (números de 0 a 9)

print(f"Shape de X (datos): {X.shape}")
print(f"Shape de y (etiquetas): {y.shape}")
print(f"Primera imagen (formato aplanado): {X[0]}")
print(f"Etiqueta correspondiente: {y[0]}")
```

```python
# Guardar algunas imagenes para usarlas como test

import matplotlib.pyplot as plt
import random

images_index = range(len(digits.images))
sample = random.sample(images_index, 5) # Se toman 5 imágenes aleatorias

for i in sample:
  plt.imshow(digits.images[i], cmap='gray')
  # plt.title(f'Etiqueta: {digits.target[0]}')
  plt.axis(False)
  plt.savefig(f'test/digit_{i}.jpg', bbox_inches='tight')
  plt.show()
```
